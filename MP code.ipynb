```python
import pandas as pd
import matplotlib.pyplot as plt 


rawDf = pd.read_csv("NBA_data.csv")
rawDf.head()
```

```python
rawDf.describe()
```

```python
def salary_formatter(x):
    return '{:,.0f}'.format(x)

# Apply the salary formatter to the 'Salary' column
rawDf['Salary'] = rawDf['Salary'].apply(salary_formatter)

# Display the modified DataFrame
print(rawDf)
```

```python
plt.figure(figsize=(8, 6))
rawDf['Position'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Player Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
rawDf['Age'].plot(kind='hist', bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
rawDf['Salary'] = rawDf['Salary'].str.replace('$', '').str.replace(',', '').astype(float)

# Define the salary ranges you want to display
salary_ranges = [(0, 1000000), (1000000, 5000000), (5000000, 10000000), (10000000, 20000000)]

# Filter the data based on the salary ranges
filtered_data = []
for salary_range in salary_ranges:
    lower_bound, upper_bound = salary_range
    filtered_data.append(rawDf[(rawDf['Salary'] >= lower_bound) & (rawDf['Salary'] < upper_bound)])

# Visualization 3: Scatter Plot - Player Salary vs. Player Age
plt.figure(figsize=(10, 6))
for idx, data_range in enumerate(filtered_data):
    plt.scatter(data_range['Salary'], data_range['Age'], label=f'Salary Range {idx+1}', alpha=0.5)
plt.title('Player Salary vs. Player Age')
plt.xlabel('Salary')
plt.ylabel('Age')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.ticklabel_format(style='plain', axis='x')  # Disable scientific notation for salary
plt.show()
```

```python
# Group the data by team and calculate the total points scored by each team
total_points_by_team = rawDf.groupby('Team')['PTS'].sum().reset_index().sort_values(by='PTS', ascending=False)

# Select only the first 30 teams
top_30_teams = total_points_by_team.head(29)

# Plot the total points scored by each team
plt.figure(figsize=(15, 6))  # Adjust the figure size for better visualization
plt.bar(top_30_teams['Team'], top_30_teams['PTS'], color='orange')
plt.title('Total Points Scored by Team (Top 30)')
plt.xlabel('Team')
plt.ylabel('Total Points')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```

```python
desired_positions = ['PG', 'SG', 'SF', 'PF', 'C']
filtered_df = rawDf[rawDf['Position'].isin(desired_positions)]

# Plot the distribution of player positions
plt.figure(figsize=(8, 8))
position_counts = filtered_df['Position'].value_counts()

plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Player Positions (PG, SG, SF, PF, C)')
plt.show()
```
